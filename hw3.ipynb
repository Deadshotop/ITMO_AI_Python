## Загрузка данных
Считываем таблицу из файла `train.csv`.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

sns.set_theme(style='whitegrid')
df = pd.read_csv('train.csv')
df.head()
```

## Распределения признаков
Показываем распределения `Survived`, `Pclass`, `Age`, `Sex`, `Parch`.

```python
fig, axes = plt.subplots(2, 3, figsize=(16, 10))
axes = axes.ravel()

survived_counts = df['Survived'].value_counts().sort_index()
axes[0].bar(survived_counts.index.astype(str), survived_counts.values, color='skyblue', label='Passengers')
axes[0].set_title('Survived Distribution')
axes[0].set_xlabel('Survived')
axes[0].set_ylabel('Count')
axes[0].legend()

pclass_counts = df['Pclass'].value_counts().sort_index()
axes[1].bar(pclass_counts.index.astype(str), pclass_counts.values, color='salmon', label='Passengers')
axes[1].set_title('Pclass Distribution')
axes[1].set_xlabel('Pclass')
axes[1].set_ylabel('Count')
axes[1].legend()

axes[2].hist(df['Age'].dropna(), bins=20, color='seagreen', edgecolor='black', label='Age')
axes[2].set_title('Age Distribution')
axes[2].set_xlabel('Age')
axes[2].set_ylabel('Count')
axes[2].legend()

sex_counts = df['Sex'].value_counts()
axes[3].bar(sex_counts.index, sex_counts.values, color='mediumpurple', label='Passengers')
axes[3].set_title('Sex Distribution')
axes[3].set_xlabel('Sex')
axes[3].set_ylabel('Count')
axes[3].legend()

parch_counts = df['Parch'].value_counts().sort_index()
axes[4].bar(parch_counts.index.astype(str), parch_counts.values, color='gold', label='Passengers')
axes[4].set_title('Parch Distribution')
axes[4].set_xlabel('Parch')
axes[4].set_ylabel('Count')
axes[4].legend()

axes[5].axis('off')
plt.tight_layout()
plt.show()
```

## Boxplot возраста
Строим boxplot для столбца `Age`

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, y='Age', color='lightblue')
plt.title('Boxplot of Age')
plt.xlabel('Passengers')
plt.ylabel('Age')
plt.legend(['Age'])
plt.show()
```

## Интерпретация boxplot
Большая часть значений возраста сосредоточена примерно до 40 лет. Видны выбросы в области старших возрастов, а медиана находится заметно ниже 40 лет

## Pie chart
Строим круговые диаграммы для `Survived` и `Pclass` с процентами

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

df['Survived'].value_counts().sort_index().plot.pie(
    autopct='%.1f%%',
    labels=['Did not survive', 'Survived'],
    ylabel='',
    ax=axes[0],
    title='Survived Pie Chart'
)
axes[0].legend(['Did not survive', 'Survived'], title='Survived', loc='upper right')

pclass_counts = df['Pclass'].value_counts().sort_index()
pclass_counts.plot.pie(
    autopct='%.1f%%',
    labels=[f'Class {value}' for value in pclass_counts.index],
    ylabel='',
    ax=axes[1],
    title='Pclass Pie Chart'
)
axes[1].legend([f'Class {value}' for value in pclass_counts.index], title='Pclass', loc='upper right')

plt.tight_layout()
plt.show()
```

## Pairplot
Строим pairplot для всех числовых признаков

```python
pairplot = sns.pairplot(df.select_dtypes(include='number').dropna(), corner=True)
pairplot.figure.suptitle('Pairplot of Numeric Features', y=1.02)
plt.show()
```

## Violin plot
Сравниваем распределение возраста по полу


```python
plt.figure(figsize=(8, 5))
sns.violinplot(data=df, x='Sex', y='Age', hue='Sex', palette='Set2')
plt.title('Age Distribution by Sex')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.legend(title='Sex')
plt.show()
```

## Heatmap
Показываем корреляционную матрицу числовых признаков

```python
plt.figure(figsize=(10, 8))
correlation = df.select_dtypes(include='number').corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.xlabel('Features')
plt.ylabel('Features')
plt.show()
```

## Sunburst plot
Строим интерактивную диаграмму по классу и полу

```python
sunburst_data = df.groupby(['Pclass', 'Sex']).size().reset_index(name='Count')
fig = px.sunburst(
    sunburst_data,
    path=['Pclass', 'Sex'],
    values='Count',
    title='Passengers by Class and Sex'
)
fig.show()
```
